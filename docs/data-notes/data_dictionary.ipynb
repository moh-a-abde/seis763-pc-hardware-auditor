```python
# import libraries

import numpy as np
import pandas as pd
```

```python
# load "computer_prices_all.csv" data
# alter if running on a different machine or if the file is in a different location

df = pd.read_csv('../seis763-pc-hardware-auditor/computer_prices_all.csv')

print(df.columns)
```

```python
# create data dictionary

data_dict = pd.DataFrame({
    "field_name": df.columns,
    "description": ["<Add description here>" for _ in df.columns],
    "data_type": [str(df[col].dtype) for col in df.columns],
    "example_value": [df[col].iloc[0] for col in df.columns],
    "distinct_values": [df[col].nunique() for col in df.columns],
    "null_count": [df[col].isna().sum() for col in df.columns],
    "notes": ["<Add notes here>" for _ in df.columns]
})

# insert description for each field
descriptions = {
    'device_type': 'Device category (either desktop or laptop)',
    'brand': 'Manufacturer brand',
    'model': 'Device model identifier (not a model number)',
    'release_year': 'Year of device release',
    'os': 'Operating system (likely pre-installed)',
    'form_factor': 'Case standards for desktop and laptops',
    'cpu_brand': 'Name of CPU vendor/manufacturer',
    'cpu_model': 'CPU model identifier (contains manufacturer name, series, and unknown numeric)',
    'cpu_tier': 'Performance tier of CPU (1 being lowest, 6 being highest)',
    'cpu_cores': 'Number of cores in CPU',
    'cpu_threads': 'Number of threads utilized by SPU',
    'cpu_base_ghz': 'Base CPU clock speeds (GHz)',
    'cpu_boost_ghz': 'Boosted CPU clock speeds (GHz)',
    'gpu_brand': 'GPU vendor (ex. NVIDIA, AMD, etc.)',
    'gpu_model': 'Name and series of GPU',
    'gpu_tier': 'Performance tier of GPU (1 being lowest, 6 being highest)',
    'vram_gb': 'Dedicated video memory (GB)',
    'ram_gb': 'System memory (GB)',
    'storage_type': 'Storage technology (ex. SSD, HDD, etc.)',
    'storage_gb': 'Storage capacity (GB)',
    'storage_drive_count': 'Number of storage devices installed',
    'display_type': 'Display technology',
    'display_size_in': 'Size of display (Inches)',
    'resolution': 'Resolution of display (pixel WidthxHeight)',
    'refresh_hz': 'Display refresh rate (Hz)',
    'battery_wh': 'Capacity of laptop battery (Wh)',
    'charger_watts': 'Wattage of laptop charger (W)',
    'psu_watts': 'Wattage of desktop power supply (W)',
    'wifi': 'Generation of Wi-Fi',
    'bluetooth': 'Version of bluetooth technology',
    'weight_kg': 'Weight of device (kg)',
    'warranty_months': 'Months of warranty issued for device',
    'price': 'Price of device'
}

data_dict['description'] = data_dict['field_name'].map(descriptions)


# add notes

notes = {
    'brand': 'sometimes has unusual connections to os (like apple brand to windows os)',
    'model': 'definitely not a traditional "model number", but maybe a unique identifier for each record',
    'os': 'sometimes has unusual connections to brand (like windows os to apple brand)',
    'form_factor': 'unsure what "mainstream" form factor is',
    'cpu_model': 'last numeric doesn\'t match existing CPU series (ex. 7700, 5800X, etc.), so it must be a value the author created',
    'gpu_model': 'likely needs cleaning (ex. data as RTX 40 60 rather than RTX4060; RX 6000 80 XT instead of RX6800XT, if that\'s what current data represents). Also, unclear if dedicated or integrated',
    'display_type': 'can infer if it\'s integrated (laptop) or external (desktop) by looking at "device_type"',
    'battery_wh': 'confirmed only for laptops',
    'charger_watts': 'confirmed only for laptops',
    'psu_watts': 'confirmed only for desktops',
    'price': 'presumably USD'
}
data_dict['notes'] = data_dict['field_name'].map(notes)

display(data_dict)
```
